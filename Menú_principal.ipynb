```python
# Ecuación para poder calcular el IMC
def calcular_imc():
  """
  Función para calcular el IMC
  La función le pide al usuario dos valores: peso y altura
    peso (float): peso del sujeto
    altura (float): altura del sujeto
    return IMC (float)
    return Clasificación del IMC (str)
  """
  peso = input ("Introduzca su peso en Kg: ")
  altura = input ("Introduzca su altura en cm: ")
  x = float (peso)
  y = float (altura)
  IMC = (x/((y/100)**2))
  ##El IMC es un float, mientras que el imc es un str, por lo que utilizar el IMC para programa futuros
  imc = f'El IMC: {round(IMC, 2)}'
  if IMC < 18.5:
    clasificacion = "Clasificación: Bajo peso"
  elif 18.5 <= IMC < 25:
    clasificacion = "Clasificación: Peso saludable"
  elif 25 <= IMC < 30:
    clasificacion = "Clasificación: Sobrepeso"
  else:
    clasificacion = "Clasificación: Obesidad"

  print ("")
  return imc, clasificacion


# Ecuación para calcular la tasa metabólica basal
def tmb():
    """
    Ecuación para calcular la Tasa Metabólica Basal con la ecuación de Roza y Shizgal de 1984
    peso (float): peso actual del sujeto en Kg
    altura (float): altura actual del sujeto m
    edad (int): edad actual del sujeto en años
    sexo (str): pregunta por el sexo biológico del sujeto
    return TMB (float): valor redondeado al segundo decimal del TMB
    """
    #validación del peso
    while True:
      peso = float (input ("Introduzca su peso en Kg: "))
      if 0 < peso <= 600:
        break
      else:
        print ("DATO NO VÁLIDO. Por favor, introduzca el dato entre 0 y 600Kg")

    #Validación de la altura
    while True:
      altura = float (input ("Introduzca su altura en cm: "))
      if 0 < altura <= 255:
        break
      else:
        print ("DATO NO VÁLIDO. Por favor, introduzca el dato entre 0 a 255cm")

    #Validación de la edad
    while True:
      edad = float (input ("Introduzca su edad actual: "))
      if 0 < edad <= 150:
        break
      else:
        print ("DATO NO VÁLIDO. Por favor, introduzca el dato entre 0 a 150 años")

    #Validación del sexo biológico
    while True:
      sexo = input ("Introduzca su sexo biológico (H/M): ").upper()
      if sexo == "H":
        #Ecuación para la Tasa Metabólica Basal para hombres
        TMB = (88.362 + (13.397 * peso) + (4.799 * altura) - (5.677 * edad))
        break

      elif sexo == "M":
        #Ecucación para la Tasa Metabólica Basal para mujeres
        TMB = (447.593 + (9.247 * peso) + (3.098 * altura) - (4.330 * edad))
        break

      else:
        print ("DATO NO VÁLIDO. Por favor, introduzca 'H o h' para Hombre o 'M o m' para Mujer")
    tmb = round(TMB, 2)
    print ("")
    return (f'Tu tasa metabólica basal es de {tmb} calorías')

#Función para llamar a las funciones IMC y Tasa Metabólica Basal
def menu_principal():
  """
  Función para llamar a las funciones de cálculo de IMC y cálculo de TMB
  """
  while True:
    print ("---SISTEMA DE ANÁLISIS BIOESTADÍSTICO---")
    print ("1. Calcular IMC")
    print ("2. Calcular TMB")
    print ("3. Salir")

    # Selección de una de las tres opciones del menú
    opcion = input ("Seleccione una de las 3 opciones (1/2/3): ")

    if opcion == "1":
      #Aquí llamamos a la ecucación del cálculo del IMC
      print (f'RESULTADO: {calcular_imc()}')
      print ("")
    elif opcion == "2":
      #Aquí llamamos a la ecuación del cálculo del TMB
      print (f'RESULTADO: {tmb()}')
      print ("")
    elif opcion == "3":
      #Aquí salimos del programa
      print (f'Saliendo del programa...')
      print ("")
      break
    else:
      print ("DATO NO VÁLIDO. Por favor, seleccione una de las 3 opciones")
      print ("")
```
